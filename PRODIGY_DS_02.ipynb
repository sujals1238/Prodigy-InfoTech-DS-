```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_data = pd.read_csv("titanic.csv")
```

```python
# Display the first few rows of the dataset
print(titanic_data.head())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())
```

```python
# Handle missing values
 
```

```python
# Fill missing age values with the median age
titanic_data['age'] = titanic_data['age'].fillna(titanic_data['age'].median())
```

```python
# Fill missing embarked values with the mode (most common value)
titanic_data['embarked'] = titanic_data['embarked'].fillna(titanic_data['embarked'].mode()[0])
```

```python
titanic_data['embark_town'] = titanic_data['embark_town'].fillna(titanic_data['embark_town'].mode()[0])
```

```python
# Drop the deck column as it has too many missing values
titanic_data = titanic_data.drop(columns=['deck'])
```

```python
# Check for missing values again
print(titanic_data.isnull().sum())
```

```python
# Convert the Survived column to a categorical type
titanic_data['survived'] = titanic_data['survived'].astype('category')
```

```python
# Convert the Pclass column to a categorical type
titanic_data['pclass'] = titanic_data['pclass'].astype('category')
```

```python
# Summary statistics for numerical variables
print(titanic_data.describe())

# Summary statistics for categorical variables
print(titanic_data.describe(include=['category']))
```

```python
# Distribution of Numerical Variables
```

```python
# Plot distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot distribution of Fare
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Categorical Variables
```

```python
# Plot count of Survived
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='survived')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Plot count of Pclass
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='Pclass')
plt.title('Passenger Class Count')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()

# Plot count of Sex
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='Sex')
plt.title('Gender Count')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
#Relationships Between Variables
```

```python
# Survival rate by gender
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic_data, x='sex', y='survived')
plt.title('Survival Rate by Gender')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate by passenger class
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic_data, x='pclass', y='survived')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate by age
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x='survived', y='age')
plt.title('Survival Rate by Age')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```
